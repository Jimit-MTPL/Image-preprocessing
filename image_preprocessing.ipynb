```python
import cv2
```

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = cv2.imread(r'C:\Users\jimit_moontechnolabs\Downloads\paris.jpeg')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis
plt.show()
```

```python
resized_image = cv2.resize(image_rgb, (224, 224))
```

```python
plt.imshow(resized_image)
plt.axis('off')
plt.show()
```

```python
height, width = resized_image.shape[:2]
size = min(height, width)
x = (width - size) // 2
y = (height - size) // 2
cropped_image = resized_image[y:y+size, x:x+size]
```

```python
plt.imshow(cropped_img)
plt.axis('off')
plt.show()
```

```python
gray_image = cv2.cvtColor(cropped_image, cv2.COLOR_RGB2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
normalized_image = gray_image / 255.0

# Display normalized image
plt.imshow(normalized_image)
plt.axis('off')
plt.show()
```

```python
equalized_image = cv2.equalizeHist(gray_image)

# Display the equalized image
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
# can use normalized image also
# Convert to unsigned 8-bit integer to properly display the image
laplacian = cv2.convertScaleAbs(laplacian)
```

```python
plt.imshow(laplacian)
plt.axis('off')
plt.show()
```

```python
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binary_image)
plt.axis('off')
plt.show()
```

```python
edges = cv2.Canny(gray_image, 100, 200)

# Display edge-detected image
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```
